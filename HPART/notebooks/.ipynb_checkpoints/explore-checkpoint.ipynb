## Start the exploration

```python
# get pandas
import pandas as pd
```

```python
# check columns with decent amount of valid data
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
# show columns with most missing entries
missing_data.head(20)
```

```python
# Now have a look at correlations with Sales price
k = 10 #number of variables for heatmap
corrmat = df_clean.corr()
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(df_clean[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
# keep cols and check also negatively correlated variables
k = 2 #number of variables for heatmap
corrmat = df_clean.corr()
ncols = corrmat.nsmallest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(df_clean[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

## Transform

```python
# check skew and kurt
for col in df_red.columns:
    print(col, df_red[col].kurt())
```

```python
# allright get into the log col those with kurt > 2
log_col = []
for col in df_red.columns:
    if abs(df_red[col].kurt()) > 2:
        log_col += [col]
print(log_col)
```

```python
# get the new transformed columns and remove the old ones
for col in log_col:
    df_red['log_' + col] = np.log(df_red[col] + 1)
    df_red = df_red.drop([col], axis=1)
print(df_red.columns)
```

```python
# time to recheck the correlation charts
sns.pairplot(df_red, size = 2.5)
```

```python
# won't do any PCA or further dim reduction at this point, simply save for the next step
df_red.to_pickle('../dataframes/df_red.pkl')  # where to save it, usually as a .pkl
```
